```python
import numpy as np
from LinearRegression import LinearRegression as LR
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
```

```python
iris
```

```python
X = iris.data
y= iris.data
```

```python
# Sepal Length: Column 0
# Sepal Width: Column 1
# Petal Length: Column 2
# Petal Width: Column 3
#their indices start from 0.
```

```python
# Model 1: Predict Sepal Length given petal length and petal width
X_train, X_test, y_train, y_test = train_test_split(X[:, 2:4], y[:, :1], test_size=0.1, random_state=0)
```

```python
# #petal width
# X_train = X[:,2:4]
# y_train = y[:,:1]
```

```python
reg_1=LR()
```

```python
reg_1.fit(X_train,y_train, regularization=0, max_epochs=100, patience=3, alpha=0.01)
```

```python
reg_1.score(X_test,y_test)
```

```python
#Model2: using the Sepal Length and Width to predict Petal Width
X_train, X_test, y_train, y_test = train_test_split(X[:, 0:2], y[:, 2:3], test_size=0.1, random_state=0)
reg_1.fit(X_train,y_train)
```

```python
reg_1.score(X_test,y_test)
```

```python
#Model-3 sepal width and petal length as the predictors and petal width as the target variable
X_train, X_test, y_train, y_test = train_test_split(X[:, 1:3], y[:, 3:4], test_size=0.1, random_state=0)
reg_1.fit(X_train,y_train)
```

```python
reg_1.score(X_train,y_train)
```

```python
#Model-4 sepal length, sepal width, and petal length as the predictors and petal width as the target variable.
X_train, X_test, y_train, y_test = train_test_split(X[:, 0:3], y[:, 3:4], test_size=0.1, random_state=0)
reg_1.fit(X_train,y_train)
```

```python
reg_1.score(X_test,y_test)
```

```python
# all the models above are with l2 regularization penalty term set to 0.
# considering MODEL#4, and adding l2 penalty term.
X_train, X_test, y_train, y_test = train_test_split(X[:, 0:3], y[:, 3:4], test_size=0.1, random_state=0)
reg_1.fit(X_train,y_train, regularization=0.1, max_epochs=100, patience=3, alpha=0.01)
```

```python
reg_1.score(X_test,y_test)
```
